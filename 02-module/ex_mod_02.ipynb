# =========================================

```python
class Car:
    def __init__(self, model, year, price):
        self.model = model
        self.year = year
        self.price = price
    
    def calculate_ipva(self):
        return self.price * 0.02
        
# Instanciando objeto
my_car = Car('Fusca', 1970, 5000)

# Acessando atributo
print(f"IPVA value: R${my_car.calculate_ipva():.2f}")
```

# =========================================
# Exercise 4 — Extensão com super
# =========================================

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

class Manager(Employee):
    def __init__(self, name, salary, bonus):
        super().__init__(name, salary)
        self.bonus = bonus

    def calculate_bonus(self):
        return self.salary * self.bonus
    
    def __str__(self):
        return (
            f"Account Manager: {self.name}\n"
            f"Salary: R${self.salary:.2f}\n"
            f"Your bonus is: R${self.calculate_bonus():.2f}"
        )
```

```python
account_manager = Manager('Alice Cardoso', 8000, 0.01)
```

```python
print(account_manager)
```

```python
class Animal:
    def speak(self):
        print("Generic Sound")

class Cat(Animal):
    def speak(self):
        print("Miauu")

class Cow(Animal):
    def speak(self):
        print("Muuuu")
```

```python
animal = Animal()
cat = Cat()
cow = Cow()
```

```python
animal.speak()
cat.speak()
cow.speak()
```

# =========================================
# Exercise 6 — Isinstance
# =========================================

```python
class Dog(Animal):
    def speak(self):
        print("Au au")
```

```python
rex = Dog()
```

```python
print(isinstance(rex, Dog))
print(isinstance(rex,Animal))
print(type(rex) == Animal)
```

`isinstance(rex, Dog)` retorna `True` porque o objeto foi criado a partir da classe `Dog`.  
`isinstance(rex, Animal)` também retorna `True`, pois `Dog` herda de `Animal`.    
Já `type(rex) == Animal` retorna `False`, porque o tipo exato do objeto é `Dog`, e não `Animal`. O operador `type()` não considera herança, apenas o tipo direto do objeto.

# Exercise 9 — Encapsulamento Lógico
# =========================================

```python
class BankAccount:
    def __init__(self):
        self.balance = 0

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
        
        else:
            print("Insufficient funds!")
```

```python
class BankAccount:
    def __init__(self):
        self.balance = 0

    def deposit(self, amount):
        if amount <= 0:
            print("Invalid amount!")
            return
        self.balance += amount

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient funds!")
            return
        self.balance -= amount
```

```python
bank_account = BankAccount()

bank_account.deposit(100)
bank_account.withdraw(50)

print(f"Account Balance: R${bank_account.balance:.2f}")
```

```python
bank_account.withdraw(90)
```

# Exercise 10 — Exceção Personalizada
# =========================================

```python
class CriticalTemperatureError(Exception):
    def __init__(self, temperature):
        self.temperature = temperature

def check_boiler(temp):
    if temp > 100:
        raise CriticalTemperatureError(temp)

try:
    check_boiler(120)
except CriticalTemperatureError as e:
    print(f"Alert! Temperature of {e.temperature}°C reached!")
```
