# =========================================
# Exercise 6 — Tratamento de Erros
# =========================================

```python
# Mean age
mean_age = df['Idade'].mean()

print(mean_age)
```

```python
# Fill missing values with mean (imputation)
df['Idade'] = df['Idade'].fillna(mean_age)
```

```python
# Inspect values before rounding
df['Idade'].unique()
```

```python
df['Idade'].value_counts().head()
```

```python
# Inspect distribution after imputation
df.describe()
```

```python
# Round and convert to integer to preserve semantic meaning of age
df['Idade'] = df['Idade'].round().astype(int)
```

```python
df['Idade'].unique()
```

```python
df['Idade'].value_counts().head()
```

```python
df.describe()
```

⚠️ **Observação sobre precisão dos dados**

Embora neste caso tenha sido realizado o arredondamento da variável `Idade`, é importante destacar que essa decisão depende do contexto dos dados.

*Caso os dados reais apresentem valores contínuos (com casas decimais), essa precisão deve ser preservada, evitando arredondamentos que possam levar à perda de informação ou distorção da distribuição original.*
