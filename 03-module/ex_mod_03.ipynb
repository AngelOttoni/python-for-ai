# =========================================
# Exercise 2 — Inspeção de Dimensões
# =========================================

```python
matrix = np.array(
    [
        [1, 2, 3, 4],
        [5, 6, 7, 8]
    ]
)
```

```python
# O formato da matriz (quantas linhas e colunas)
print(f"Shape: {matrix.shape}")

# O número total de elementos na matriz
print(f"Total elements: {matrix.size}")

# O número de dimensões (eixos) da matriz
print(f"Number of dimensions: {matrix.ndim}")
```

# =========================================
# Exercise 3 — Matemática Vetorizada
# =========================================

```python
values = np.array([
    50, 100, 150, 200
])

print(values)

# Aplica aumento de 10% a todos os valores simultaneamente (multiplique por 1.1)
values = values * 1.1

#Subtraia 5 de todos os valores obtidos
values = values - 5

# Imprima o array resultante
print(values)
```

# =========================================
# Exercise 5 — Produto Matricial
# =========================================

```python
A = np.array([
    [1, 2],
    [3, 4]
])
```

```python
B = np.array([
    [5 ,6],
    [7, 8]
])
```

```python
result = A @ B
print(result)
```

**Conceito importante** 
- \* → multiplicação elemento a elemento   
- @ → multiplicação matricial real

# =========================================
# Exercise 6 — Estatísticas
# =========================================

```python
grades = np.array([7.5, 8.0, 9.5, 6.0, 10.0])

print("Max:", np.max(grades))
print("Min:", np.min(grades))
print("Mean:", np.mean(grades))
```

# =========================================
# Exercise 7 — A função `argmax`
# =========================================

```python
probs = np.array([0.15, 0.75, 0.10])

predicted_class = np.argmax(probs)

print(f"Predicted class index: {predicted_class}")
```

# =========================================
# Exercise 8 — Reshape e Flatten
# =========================================

```python
vector = np.arange(12)

print(vector)
```

```python
matrix = vector.reshape(3, 4)

print(matrix)
```

```python
flattened = matrix.flatten()

print(flattened)
```

# =========================================
# Exercise 9 — Filtros Lógicos
# =========================================

```python
ages = np.array([12, 18, 25, 15, 30, 42, 17])

adult_ages = ages[ages >= 18]

print(adult_ages)
```

# =========================================
# Exercise 10 — Limpeza Condicional
# =========================================

```python
temp = np.array([30, -10, 25, 28, -5])

clean_temp = np.where(temp < 0, 0, temp)

print(clean_temp)
```
