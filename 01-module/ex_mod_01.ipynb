# Exercise 3 — Gerenciamento de Listas
# =========================================

```python
grades = [5.0, 10.0, 6.5]

# (a) Add 8.0 to the end
grades.append(8.0)

# (b) Insert 9.0 at position 0
grades.insert(0, 9.0)

# (c) Remove 5.0
grades.remove(5.0)

# (d) Print final list
print("Final grades list:", grades)
```

```python
grade = float(input("Enter the student's grade: "))

if grade >= 7:
    print("Approved")
elif grade >= 5:
    print("Recovery")
else:
    print("Failed")
```

# =========================================
# Exercise 8 — Modularização
# =========================================

```python
def calculate_circle_area(radius):
    area = 3.14 * (radius ** 2)
    return area

print("Circle area:", calculate_circle_area(5))
```
